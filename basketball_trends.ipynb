```python
def myplot(playerlist):
    Col = {"KobeBryant":'Green',"JoeJohnson":'Black',"LeBronJames":'Blue',"CarmeloAnthony":'red',"DwightHoward":'pink',"ChrisBosh":'magenta',"ChrisPaul":'Brown',"KevinDurant":'Cyan',"DerrickRose":'grey',"DwayneWade":'yellow'}
    mar = {"KobeBryant":'.',"JoeJohnson":'o',"LeBronJames":',',"CarmeloAnthony":'v',"DwightHoward":'^',"ChrisBosh":'p',"ChrisPaul":'P',"KevinDurant":'h',"DerrickRose":'x',"DwayneWade":'|'}
    for name in playerlist:
        plt.plot(Games[Pdict[name]], c=Col[name], ls='--', marker=mar[name], ms=5, label = name)
    plt.legend(loc = 'lower left', bbox_to_anchor = (1,0))
    plt.xticks(list(range(0,9)),Seasons, rotation = 'vertical')
    plt.show()
 
```

```python
def myplot(data, playerlist = Players):
    Col = {"KobeBryant":'Green',"JoeJohnson":'Black',"LeBronJames":'Blue',"CarmeloAnthony":'red',"DwightHoward":'pink',"ChrisBosh":'magenta',"ChrisPaul":'Brown',"KevinDurant":'Cyan',"DerrickRose":'grey',"DwayneWade":'yellow'}
    mar = {"KobeBryant":'.',"JoeJohnson":'o',"LeBronJames":',',"CarmeloAnthony":'v',"DwightHoward":'^',"ChrisBosh":'p',"ChrisPaul":'P',"KevinDurant":'h',"DerrickRose":'x',"DwayneWade":'|'}
    for name in playerlist:
        plt.plot(data[Pdict[name]], c=Col[name], ls='--', marker=mar[name], ms=5, label = name)
    plt.legend(loc = 'lower left', bbox_to_anchor = (1,0))
    plt.xticks(list(range(0,9)),Seasons, rotation = 'vertical')
    plt.show()
 
```
